# Observando as perguntas de cada DataFrame

```python
def get_questions(df):
    questions = df['question_label'].unique()
    for i, q in enumerate(questions):
        print('-> Question {}: {}'.format(i+1, q))
```

## Perguntas do DailyLife

```python
get_questions(DailyLife)
```

## Perguntas do Discrimination

```python
get_questions(Discrim)
```

## Perguntas do RightsAwareness

```python
get_questions(RightsAw)
```

## Perguntas do TransgenderSpecificQuestions

```python
get_questions(TransSpec)
```

## Perguntas do ViolenceAndHarassment
